```python
import pandas as pd
import numpy as np
import os
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
#%matplotlib notebook
%matplotlib inline
```

```python
classical_datasets_path = "adbench/datasets/Classical/{}"

all_classical_datasets = os.listdir(classical_datasets_path.format(""))
```

```python
details_classical_datasets = []

for dataset in all_classical_datasets:
    data = np.load(classical_datasets_path.format(dataset), allow_pickle=True)
    X, y = data['X'], data['y']
    parts = dataset.split("_")
    details_classical_datasets.append([dataset, parts[0], parts[1],X.shape[0], X.shape[1], y.sum(), y.sum()/y.shape[0]])

details_classical_datasets = pd.DataFrame(details_classical_datasets, columns=["Filename", "Number", "Dataset", "Samples", "Features", "Anomaly", "%Anomaly"])
details_classical_datasets["Number"] = details_classical_datasets["Number"].astype(int)
details_classical_datasets = details_classical_datasets.sort_values("Number").reset_index(drop=True)
details_classical_datasets
```

```python
datasets_path = "adbench/datasets/CV_by_ResNet18/{}"

all_datasets = os.listdir(datasets_path.format(""))
```

```python
details = []

for dataset in all_datasets:
    data = np.load(datasets_path.format(dataset), allow_pickle=True)
    X, y = data['X'], data['y']
    details.append([dataset,X.shape[0], X.shape[1], y.sum(), y.sum()/y.shape[0]])

details = pd.DataFrame(details, columns=["Filename",  "Samples", "Features", "Anomaly", "%Anomaly"])
details
```

# Dataframe for categories and their occurrences

```python
category = df['Category'].value_counts().reset_index()
category.columns = ['Category', 'Count']
```

# Category

```python
fig, ax = plt.subplots(figsize=(10,6))

sns.barplot(data=category, y='Category', x='Count', hue='Category', palette='viridis')
plt.xlabel('Count')
plt.ylabel('Category')
plt.title("Dataset by Category")
#plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

plt.pie(category['Count'], labels=category['Category'], autopct='%.0f%%')

plt.title("% Dataset by Category")
# plt.legend()
plt.show()
```

# ECDF (Empirical Cumulative Distribution Function) for # sample

```python
percentage_list = np.arange(0,1,0.01)

df_sorted_features = df.sort_values(by="Samples") 

sample_quantile = df_sorted_features.Samples.quantile(percentage_list)

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(sample_quantile, percentage_list)
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, 500000, step=50000), rotation=20)
plt.xlabel('# Samples')
plt.ylabel('ECDF')
plt.title("ECDF for # Samples")
# plt.legend()
plt.show()
```

# Features

```python
df_sorted_features = df.sort_values('Features', ascending=True)
df_sorted_features

fig, ax = plt.subplots(figsize=(15,12))

sns.barplot(data=df_sorted_features, x='Features', y='Data', hue='Data', palette='viridis')
plt.xticks([10, 100, 500, 800, 1500])
plt.xlabel('Number of features')
plt.ylabel('Dataset name')
plt.title("Datasets vs # Features")
# plt.legend()
plt.show()
```

# ECDF (Empirical Cumulative Distribution Function) for # features

```python
percentage_feature = np.arange(0,1,0.01)

feature_quantile = df_sorted_features.Features.quantile(percentage_feature)

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(feature_quantile, percentage_feature)
plt.yticks(np.arange(0, 1.1, step=0.1))
plt.xticks(np.arange(0, 1100, step=100))
plt.xlabel('# Features')
plt.ylabel('ECDF')
plt.title("ECDF for # Features")
# plt.legend()
plt.show()
```

# ECDF (Empirical Cumulative Distribution Function) for % Anomaly

```python
percentage_feature = np.arange(0,1,0.01)

per_anomaly_quantile = df_sorted_features.per_Anamaly.quantile(percentage_feature)

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(per_anomaly_quantile, percentage_feature)
plt.yticks(np.arange(0, 1.1, step=0.1))
# plt.xticks(np.arange(0, 1100, step=100))
plt.xlabel('% Anomaly')
plt.ylabel('ECDF')
plt.title("ECDF for % Anomaly")
# plt.legend()
plt.show()
```
